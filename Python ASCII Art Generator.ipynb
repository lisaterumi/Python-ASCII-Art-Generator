## Geração de imagem em ASCII

Adaptado de [python-ascii-art-generator](https://medium.com/javascript-dots/python-ascii-art-generator-60ba9eb559d7).

Passos para converter a imagem em caracteres ASCII:

1. Carregar a imagem
1. Redimensionar a imagem
1. Converter imagem em escala de cinza
1. Converta cada pixel no respectivo caractere ASCII

### Carregando uma imagem usando a biblioteca de imagens PIL

Para carregar a imagem, usaremos a biblioteca PIL.

````python
import PIL.Image

def main():
    path = input("Entre com o caminho da imagem: \n")
    try:
        image = PIL.Image.open(path)
    except:
        print(path, "Não foi possível encontrar a imagem. ");
````

O código acima lê a imagem a partir do caminho fornecido pelo usuário. Se a imagem não existir no caminho fornecido, mostraremos uma mensagem de erro.

### Definir lista ASCII

Vamos criar uma lista de caracteres ASCII:

````python
ASCII_CHARS = ["@", "#", "$", "%", "?", "*", "+", ";", ":", ",", "."]
````

Os caracteres ASCII são organizados do mais escuro ao mais claro. Isso significa que o pixel mais escuro será substituído por ```@``` e o mais claro por ```.```.

### Redimensionar imagem

Precisamos redimensionar a imagem, para que não resulte em um texto de tamanho grande. Para encontrar ```new_height```, multiplique ```new_width``` por ```old_height``` e então divida por ```old_width```.

````python
def resize(image, new_width = 100):
    width, height = image.size
    new_height = new_width * height / width
    return image.resize((int(new_width), int(new_height)))
````

### Converter imagem para escala de cinza

Podemos usar o método de conversão na imagem com a opção ```L``` para obter a imagem em *GrayScale*.

````python
def to_greyscale(image):
    return image.convert("L")
````

### Converter imagem em caractere ASCII

Para converter a imagem em caracteres ASCII, vamos obter o valor de cada pixel (0-255). Obtemos o caractere ASCII correspondente e juntamos as strings.

````python
def pixel_to_ascii(image):
    pixels = image.getdata()
    ascii_str = "";
    for pixel in pixels:
        ascii_str += ASCII_CHARS[pixel//25];
    return ascii_str
````

Agora temos um método ```to_greyscale``` que converte imagem em imagem *GreyScale*, método ```pixel_to_ascii``` para converter imagem em string ASCII, e uma vez que obtemos a string ASCII da imagem, precisamos dividir a string com base na largura da imagem e salvá-la em um Arquivo.

````python
import PIL.Image
def main():
    path = input("Entre com o caminho da imagem : \n")
    try:
        image = PIL.Image.open(path)
    except:
        print(path, "Não foi possível encontrar a imagem. ")
    #redimensiona image
    image = resize(image);
    #converte para escala de cinza 
    greyscale_image = to_greyscale(image)
    #converte para caracteres ascii 
    ascii_str = pixel_to_ascii(greyscale_image)
    img_width = greyscale_image.width
    ascii_str_len = len(ascii_str)
    ascii_img=""
    #Divide a string com base no tamanho da imagem
    for i in range(0, ascii_str_len, img_width):
        ascii_img += ascii_str[i:i+img_width] + "\n"
    #salva a string em um arquivo
    with open("ascii_image.txt", "w") as f:
        f.write(ascii_img);
main()
````
